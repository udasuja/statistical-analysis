```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%precision 3
pd.set_option('display.precision',3)
```

```python
df=pd.read_csv('/content/python_stat_sample-master/data/ch2_scores_em.csv',index_col='student number')
```

```python
en_scores=np.array(df['english'])[:10]
ma_scores=np.array(df['mathematics'])[:10]

scores_df=pd.DataFrame({'english':en_scores,'mathematics':ma_scores},
                       index=pd.Index(['A','B','C','D','E','F','G','H','I','J'],name='student'))
scores_df
```

```python
summary_df=scores_df.copy()
summary_df['english_deviation']=summary_df['english']-summary_df['english'].mean()
summary_df['mathematics_deviation']=summary_df['mathematics']-summary_df['mathematics'].mean()
summary_df['product of deviations']=summary_df['english_deviation']*summary_df['mathematics_deviation']
summary_df
```

```python
print(summary_df['product of deviations'].mean())
print("수학 평균",summary_df['mathematics'].mean())
print("영어 평균",summary_df['english'].mean())
```

```python
cov_mat=np.cov(en_scores,ma_scores,ddof=0)
print("공분산 행렬:\n",cov_mat)
print("공분산 행렬의 [0,1]성분:{}\n공분산 행렬의 [1,0]성분:{}".format(cov_mat[0,1],cov_mat[1,0]))
print("공분산 행렬의 [0,0]성분:{}\n공분산 행렬의 [1,1]성분:{}".format(cov_mat[0,0],cov_mat[1,1]))
print("영어 분산",np.var(en_scores,ddof=0))
print("수학 분산",np.var(ma_scores,ddof=0))
```

```python
print(np.cov(en_scores,ma_scores,ddof=0)[0,1]/(np.std(en_scores)*np.std(ma_scores)))
```

```python
cor_coe=np.corrcoef(en_scores,ma_scores)
print(cor_coe)
print("상관계수 행렬의 [0,0]성분:{}".format(cor_coe[0,0]))
print("상관계수 행렬의 [0,1]성분:{}".format(cor_coe[0,1]))
print("상관계수 행렬의 [1,0]성분:{}".format(cor_coe[1,0]))
print("상관계수 행렬의 [1,1]성분:{}".format(cor_coe[1,1]))
print("영어와 영어의 상관계수:{}".format(np.cov(en_scores,en_scores,ddof=0)[0,1]/(np.std(en_scores)*np.std(en_scores))))
print("수학과 수학의 상관계수:{}".format(np.cov(ma_scores,ma_scores,ddof=0)[0,1]/(np.std(ma_scores)*np.std(ma_scores))))
```

```python
english_scores=np.array(df['english'])
math_scores=np.array(df['mathematics'])

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111)

ax.scatter(english_scores,math_scores)
ax.set_xlabel('english')
ax.set_ylabel('mathematics')

plt.show()
```

```python
poly_fit=np.polyfit(english_scores,math_scores,1)
poly_1d=np.poly1d(poly_fit)
xs=np.linspace(english_scores.min(),english_scores.max())
ys=poly_1d(xs)

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111)
ax.plot(xs,ys,color='gray',label=f'{poly_fit[1]:.2f}+{poly_fit[0]:.2f}x')
ax.scatter(english_scores,math_scores,label='score')
ax.set_xlabel('english')
ax.set_ylabel('mathematics')
ax.legend(loc='upper left')

plt.show()
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)

c=ax.hist2d(english_scores,math_scores,bins=[9,8],range=[(35,80),(55,95)])
ax.set_xlabel('english')
ax.set_ylabel('mathematics')
ax.set_xticks(c[1])
ax.set_yticks(c[2])

fig.colorbar(c[3],ax=ax)
plt.show()
```

```python
anscombe_data=np.load('/content/python_stat_sample-master/data/ch3_anscombe.npy')
print(anscombe_data.shape)
anscombe_data[0]
```

```python
stats_df=pd.DataFrame(index=['X_mean','X_variance','Y_mean','Y_variance','X&Y_correlation','X&Y_regression line'])
for i, data in enumerate(anscombe_data):
  dataX=data[:,0]
  dataY=data[:,1]
  poly_fit=np.polyfit(dataX,dataY,1)
  stats_df[f'data{i+1}']=\
    [f'{np.mean(dataX):.2f}',
     f'{np.var(dataX):.2f}',
     f'{np.mean(dataY):.2f}',
     f'{np.var(dataY):.2f}',
     f'{np.corrcoef(dataX,dataY)[0,1]:.2f}',
     f'{poly_fit[1]:.2f}+{poly_fit[0]:.2f}x'
     ]
stats_df
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(10,10),sharex=True,sharey=True)
xs=np.linspace(0,30,100)
for i, data in enumerate(anscombe_data):
  poly_fit=np.polyfit(data[:,0],data[:,1],1)
  poly_1d=np.poly1d(poly_fit)
  ys=poly_1d(xs)
  ax=axes[i//2,i%2]
  ax.set_xlim([4,20])
  ax.set_ylim([3,13])
  ax.set_title(f'data{i+1}')
  ax.scatter(data[:,0],data[:,1])
  ax.plot(xs,ys,color='gray')

plt.tight_layout()
plt.show()
```
